```python
import pandas as pd
import matplotlib.pyplot as plt

products = {"分類": ["Home","Home","Entertainment","Entertainment","Technology","Technology"],
       "商店": ["Carrefour A", "RT-Mart A", "Carrefour B", "Quanlianchao A", "RT-Mart B", "Carrefour C"],
       "價格": [11.42,23.50,19.99,15.95,55.75,111.55]}

df = pd.DataFrame(products, columns = ["分類", "商店", "價格"])
df["價格"].plot(kind='line') # 繪製折線圖
plt.grid(True) # 顯示圖表的格線

plt.xticks([0, 1, 2, 3, 4, 5], df["商店"])
plt.xlabel("Name")
plt.ylabel("Price")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

products = {"分類": ["Home","Home","Entertainment","Entertainment","Technology","Technology"],
       "商店": ["Carrefour A", "RT-Mart A", "Carrefour B", "Quanlianchao A", "RT-Mart B", "Carrefour C"],
       "價格": [11.42,23.50,19.99,15.95,55.75,111.55]}

df = pd.DataFrame(products, columns = ["分類", "商店", "價格"])
df["價格"].plot(kind='line') # 繪製折線圖
plt.grid(True) # 顯示圖表的格線

df_grouped = df.groupby("分類")
type = df_grouped["商店"].count()
type.plot(kind='pie', title="Name")
plt.ylabel("")
plt.show()
```
